```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 函数的定义极其灵活
def f(x, a=1, b=1):
    return a + b * x
print(f(2))
print(f(2, a=2, b=2))
```

```python
# 函数的紧凑形式
f = lambda x: x**3
f(3)
```

```python
# 不必给函数定义名字

from scipy.integrate import quad

quad(lambda x: x**3, 0, 2)
```

```python
def generate_data(n, generator_type):
    ϵ_values = []
    for i in range(n):
        if generator_type == 'U':
            e = np.random.uniform(0, 1)
        else:
            e = np.random.randn()
        ϵ_values.append(e)
    return ϵ_values

data = generate_data(100, 'U')
plt.plot(data)
plt.show()
```

```python
def x(t):
    if t == 0:
        return 1
    else:
        return 2 * x(t-1)

x(10)
```

### 阶乘

```python
# recursively defined
def factorial(n):
    if n == 0:
        return 1
    else:
        return n*factorial(n-1)

factorial(4)
```

```python
from numpy.random import uniform

def binomial_rv(n,p):
    count = 0
    for i in range(n):
        u = uniform()
        if u < p:
            count += 1
    return count

binomial_rv(10,0.5)
```
